```python
df.hist(figsize=(9, 8), bins=20, color='skyblue', edgecolor='black')
plt.suptitle('Distribution of Numerical Columns', fontsize=16)
plt.show()

plt.figure(figsize=(6, 3))
sns.boxplot(data=df.select_dtypes(include=np.number))
plt.title('Boxplot of Numerical Columns', fontsize=11)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', data=df, palette='coolwarm')
plt.title('Gender Distribution', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(y='Job Title', data=df, palette='Set2')
plt.title('Job Title Distribution', fontsize=14)
plt.show()
```

```python
# Convert 'Date of birth' to datetime
df['Date of birth'] = pd.to_datetime(df['Date of birth'])

# Extract year of birth for analysis
df['Year of birth'] = df['Date of birth'].dt.year

# Plot histogram of Year of birth
plt.figure(figsize=(6,4))
sns.histplot(df['Year of birth'], bins=10, color='skyblue', kde=True)
plt.title('Distribution of Year of Birth', fontsize=14)
plt.show()
```

```python
# Display the unique job titles
print("\nUnique Job Titles:")
print(df['Job Title'].unique())

# Analyze the length of phone numbers
df['Phone Length'] = df['Phone'].apply(lambda x: len(x))

# Display phone number length statistics
print("\nPhone Number Length Statistics:")
print(df['Phone Length'].describe())

# Plot the distribution of phone number lengths
plt.figure(figsize=(6,4))
sns.histplot(df['Phone Length'], bins=5, color='green', kde=True)
plt.title('Distribution of Phone Number Length', fontsize=14)
plt.show()
```

```python
# Extract email domains
df['Email Domain'] = df['Email'].apply(lambda x: x.split('@')[1])

# Count plot for Email domains
plt.figure(figsize=(8,6))
sns.countplot(y='Email Domain', data=df, palette='husl')
plt.title('Email Domain Distribution', fontsize=14)
plt.show()
```
